```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import pennylane as qml
import numpy as np

# ==========================================
# 1. Quantum Topology Setup
# ==========================================
n_qubits = 4
q_layers = 3

# Native simulators for rapid gradient computation
dev1 = qml.device("default.qubit", wires=n_qubits)
dev2 = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev1, interface="torch")
def quantum_vqc(inputs, weights):
    qml.templates.AngleEmbedding(inputs, wires=range(n_qubits), rotation='X')
    qml.templates.StronglyEntanglingLayers(weights, wires=range(n_qubits))
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

@qml.qnode(dev2, interface="torch")
def quantum_vqc2(inputs, weights):
    qml.templates.AngleEmbedding(inputs, wires=range(n_qubits), rotation='X')
    qml.templates.StronglyEntanglingLayers(weights, wires=range(n_qubits))
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]


# ==========================================
# 2. SOTA CLASSICAL COMPONENTS
# ==========================================
class GatedResidualBlock(nn.Module):
    def __init__(self, input_dim, output_dim, dropout=0.1):
        super().__init__()
        self.linear1 = nn.Linear(input_dim, output_dim)
        self.linear2 = nn.Linear(output_dim, output_dim)
        self.gate = nn.Linear(output_dim, output_dim)
        self.dropout = nn.Dropout(dropout)

        # Match dimensions for skip connection if needed
        self.skip = nn.Linear(input_dim, output_dim) if input_dim != output_dim else nn.Identity()

    def forward(self, x):
        residual = self.skip(x)
        x = torch.relu(self.linear1(x))
        x = self.dropout(x)
        x = self.linear2(x)
        # Gated Linear Unit activation mechanism
        x = x * torch.sigmoid(self.gate(x))
        return x + residual


# ==========================================
# 3. ENTIRE HYBRID NETWORK ARCHITECTURE
# ==========================================
class EndToEndParallelQNN(nn.Module):
    def __init__(self, input_dim=11, output_dim=1):
        super().__init__()

        # Classical Front-End: Process and condense incoming water telemetry
        self.encoder = nn.Sequential(
            GatedResidualBlock(input_dim, 32),
            GatedResidualBlock(32, 16)
        )

        # Mid-splitting: Map representation vector to two distinct quantum feature-heads
        self.to_q1 = nn.Linear(16, n_qubits)
        self.to_q2 = nn.Linear(16, n_qubits)

        # PennyLane Quantum Layers wrapped for PyTorch autograd
        weight_shapes = {"weights": (q_layers, n_qubits, 3)}
        self.qlayer1 = qml.qnn.TorchLayer(quantum_vqc, weight_shapes)
        self.qlayer2 = qml.qnn.TorchLayer(quantum_vqc2, weight_shapes)

        # Classical Back-End: Fuse the expectation values from both circuits
        self.decoder = nn.Sequential(
            GatedResidualBlock(n_qubits * 2, 16),
            nn.Linear(16, output_dim)
        )

    def forward(self, x):
        # Step 1: Classical encoding
        features = self.encoder(x)

        # Step 2: Route features into distinct quantum heads, bounded to [-pi, pi]
        q1_in = torch.tanh(self.to_q1(features)) * np.pi
        q2_in = torch.tanh(self.to_q2(features)) * np.pi

        # Step 3: Run Quantum Circuits
        q1_out = self.qlayer1(q1_in)
        q2_out = self.qlayer2(q2_in)

        # Step 4: Concatenate quantum outputs (Shape: Batch, n_qubits * 2)
        combined_quantum_states = torch.cat((q1_out, q2_out), dim=1)

        # Step 5: Final output regression mapping (PT-100)
        return self.decoder(combined_quantum_states)


# Build standard PyTorch DataLoaders
train_dataset = TensorDataset(X_train_t, y_train_t)
test_dataset = TensorDataset(X_test_t, y_test_t)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)


# ==========================================
# 5. EXECUTION: THE TRAIN & EVALUATION LOOP
# ==========================================
# The input_dim should match the number of features in X_train_t
model = EndToEndParallelQNN(input_dim=X_train_t.shape[1], output_dim=1)
criterion = nn.MSELoss()
optimizer = optim.AdamW(model.parameters(), lr=0.01, weight_decay=0.01)

epochs = 5
print("🚀 Starting Hybrid Quantum-Classical Training Loop...\n")

for epoch in range(epochs):
    model.train()
    running_loss = 0.0

    for batch_X, batch_y in train_loader:
        optimizer.zero_grad()

        # Forward pass through whole pipeline
        predictions = model(batch_X)
        loss = criterion(predictions, batch_y)

        # Backpropagate gradients across classical and quantum nodes
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * batch_X.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    print(f"Epoch [{epoch+1}/{epochs}] - Training MSE Loss: {epoch_loss:.5f}")

# ==========================================
# 6. MODEL VALIDATION / TESTING PHASE
# ==========================================
model.eval()
test_loss = 0.0

print("\n Entering Validation Phase...")
with torch.no_grad():
    for batch_X, batch_y in test_loader:
        test_preds = model(batch_X)
        loss = criterion(test_preds, batch_y)
        test_loss += loss.item() * batch_X.size(0)

final_test_mse = test_loss / len(test_loader.dataset)
print(f" Final Out-of-Sample Test MSE: {final_test_mse:.5f}")
print("\n Pipeline execution complete! Ready to swap in your real CSV tensors.")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import pennylane as qml
import numpy as np

# ==========================================
# 1. STABILIZED QUANTUM TOPOLOGY
# ==========================================
n_qubits = 4
q_layers = 2

dev1 = qml.device("default.qubit", wires=n_qubits)
dev2 = qml.device("default.qubit", wires=n_qubits)

def vqc_base(inputs, weights):
    # Initial data angle mapping
    qml.templates.AngleEmbedding(inputs, wires=range(n_qubits), rotation='Y')

    # First variational entangling layer block
    qml.templates.StronglyEntanglingLayers(weights[0], wires=range(n_qubits))

    # Internal mid-circuit entanglement routing
    qml.CNOT(wires=[0, 1])
    qml.CNOT(wires=[1, 2])
    qml.CNOT(wires=[2, 3])

    # Final variational processing layer block
    qml.templates.StronglyEntanglingLayers(weights[1], wires=range(n_qubits))

    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

qnode1 = qml.QNode(vqc_base, dev1, interface="torch")
qnode2 = qml.QNode(vqc_base, dev2, interface="torch")


# ==========================================
# 2. SOTA ATTENTION & RESIDUAL BLOCKS
# ==========================================
class HydroFeatureAttention(nn.Module):
    """Computes self-attention across tabular water features to emphasize anomalies."""
    def __init__(self, num_features, head_dim=1):
        super().__init__()
        self.num_features = num_features
        self.head_dim = head_dim

        self.query_proj = nn.Linear(1, head_dim)
        self.key_proj   = nn.Linear(1, head_dim)
        self.value_proj = nn.Linear(1, head_dim)
        self.scale = np.sqrt(head_dim)
        self.output_proj = nn.Linear(head_dim, 1)

    def forward(self, x):
        x_reshaped = x.unsqueeze(-1) # (batch_size, num_features, 1)

        q = self.query_proj(x_reshaped)
        k = self.key_proj(x_reshaped)
        v = self.value_proj(x_reshaped)

        attn_scores = torch.softmax((q @ k.transpose(-2, -1)) / self.scale, dim=-1)
        attention_output = attn_scores @ v

        output = self.output_proj(attention_output).squeeze(-1)
        return output


class GatedResidualBlock(nn.Module):
    def __init__(self, input_dim, output_dim, dropout=0.3):
        super().__init__()
        self.linear1 = nn.Linear(input_dim, output_dim)
        self.linear2 = nn.Linear(output_dim, output_dim)
        self.gate = nn.Linear(output_dim, output_dim)
        self.dropout = nn.Dropout(dropout)

        # Using exact nn.Identity mapping for uniform feature dimension paths
        self.skip = nn.Identity() if input_dim == output_dim else nn.Linear(input_dim, output_dim)

    def forward(self, x):
        residual = self.skip(x)
        x_act = torch.relu(self.linear1(x))
        x_act = self.dropout(x_act)
        x_act = self.linear2(x_act)
        x_act = x_act * torch.sigmoid(self.gate(x_act))
        return x_act + residual


# ==========================================
# 3. END-TO-END ADVANCED CLASSIFIER ARCHITECTURE
# ==========================================
class QuantumAttentionInfrastructureGuard(nn.Module):
    def __init__(self, input_dim):
        super().__init__()

        self.attention = HydroFeatureAttention(num_features=input_dim)

        self.encoder = nn.Sequential(
            GatedResidualBlock(input_dim, 32, dropout=0.3),
            GatedResidualBlock(32, 16, dropout=0.3)
        )

        self.to_q1 = nn.Linear(16, n_qubits)
        self.to_q2 = nn.Linear(16, n_qubits)
        self.quantum_dropout = nn.Dropout(p=0.2)

        # Correct shape assignment for our custom two-tier block circuit weights
        # Shape: (2 blocks, q_layers, n_qubits, 3 rotation dimensions)
        weight_shapes = {"weights": (2, q_layers, n_qubits, 3)}
        self.qlayer1 = qml.qnn.TorchLayer(qnode1, weight_shapes)
        self.qlayer2 = qml.qnn.TorchLayer(qnode2, weight_shapes)

        # STABLE UPGRADE: Removed Sigmoid() completely from the final layer block
        self.decoder = nn.Sequential(
            GatedResidualBlock(n_qubits * 2, 16, dropout=0.2),
            nn.Linear(16, 1)
        )

    def forward(self, x):
        x = self.attention(x)
        features = self.encoder(x)

        q1_in = torch.tanh(self.to_q1(features)) * np.pi
        q2_in = torch.tanh(self.to_q2(features)) * np.pi

        q1_in = self.quantum_dropout(q1_in)
        q2_in = self.quantum_dropout(q2_in)

        q1_out = self.qlayer1(q1_in)
        q2_out = self.qlayer2(q2_in)

        combined_quantum_signals = torch.cat((q1_out, q2_out), dim=1)
        return self.decoder(combined_quantum_signals)


# ==========================================
# 4. DATA LOADER PREPARATION
# ==========================================
input_features_count = X_train_raw.shape[1]


# ==========================================
# 5. EXECUTION: TWO-SPEED STABILIZED OPTIMIZER
# ==========================================
model = QuantumAttentionInfrastructureGuard(input_dim=input_features_count)

# HIGH STABILITY LOSS HEADER
criterion = nn.BCEWithLogitsLoss()

classical_params = []
quantum_params = []

for name, param in model.named_parameters():
    if "quantum_layer" in name or "qlayer" in name:
        quantum_params.append(param)
    else:
        classical_params.append(param)

optimizer = optim.AdamW([
    {'params': classical_params, 'lr': 0.005, 'weight_decay': 0.01},
    {'params': quantum_params, 'lr': 0.0005, 'weight_decay': 0.001}
])

epochs = 5
print("🚀 Launching Logit-Stabilized Quantum-Attention Training Loop...\n")

for epoch in range(epochs):
    model.train()
    running_loss = 0.0

    for batch_X, batch_y in train_loader:
        optimizer.zero_grad()

        # Predictions here output raw unbounded logits
        predictions = model(batch_X)
        # Reshape batch_y to match predictions [batch_size, 1]
        loss = criterion(predictions, batch_y.unsqueeze(1).float())

        loss.backward()
        optimizer.step()

        running_loss += loss.item() * batch_X.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    print(f"Epoch [{epoch+1}/{epochs}] - Training Logit BCE Loss: {epoch_loss:.5f}")


# ==========================================
# 6. MODEL VALIDATION / EVALUATION
# ==========================================
model.eval()
test_loss = 0.0
correct_predictions = 0

print("\n🔒 Entering Validation Phase...")
with torch.no_grad():
    for batch_X, batch_y in test_loader:
        test_preds = model(batch_X)
        # Reshape batch_y to match predictions [batch_size, 1]
        loss = criterion(test_preds, batch_y.unsqueeze(1).float())
        test_loss += loss.item() * batch_X.size(0)

        # STABLE ACCURACY CHECK: Explicitly computing the Sigmoid boundary mapping
        binary_preds = (torch.sigmoid(test_preds) >= 0.5).float()
        correct_predictions += (binary_preds == batch_y.unsqueeze(1)).sum().item()

final_test_bce = test_loss / len(test_loader.dataset)
final_accuracy = (correct_predictions / len(test_loader.dataset)) * 100

print(f"➡️ Final Out-of-Sample Test Logit BCE Loss: {final_test_bce:.5f}")
print(f"➡️ Total Classification Accuracy: {final_accuracy:.2f}%")
print("\n✨ Process finished cleanly. Run this exact setup for Leak and Burst instances.")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import pennylane as qml
import numpy as np

# ==========================================
# 1. STABILIZED QUANTUM TOPOLOGY
# ==========================================
n_qubits = 4
q_layers = 2

dev1 = qml.device("default.qubit", wires=n_qubits)
dev2 = qml.device("default.qubit", wires=n_qubits)

def vqc_base(inputs, weights):
    # Initial data angle mapping
    qml.templates.AngleEmbedding(inputs, wires=range(n_qubits), rotation='Y')

    # First variational entangling layer block
    qml.templates.StronglyEntanglingLayers(weights[0], wires=range(n_qubits))

    # Internal mid-circuit entanglement routing
    qml.CNOT(wires=[0, 1])
    qml.CNOT(wires=[1, 2])
    qml.CNOT(wires=[2, 3])

    # Final variational processing layer block
    qml.templates.StronglyEntanglingLayers(weights[1], wires=range(n_qubits))

    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

qnode1 = qml.QNode(vqc_base, dev1, interface="torch")
qnode2 = qml.QNode(vqc_base, dev2, interface="torch")


# ==========================================
# 2. SOTA ATTENTION & RESIDUAL BLOCKS
# ==========================================
class HydroFeatureAttention(nn.Module):
    """Computes self-attention across tabular water features to emphasize anomalies."""
    def __init__(self, num_features, head_dim=1):
        super().__init__()
        self.num_features = num_features
        self.head_dim = head_dim

        self.query_proj = nn.Linear(1, head_dim)
        self.key_proj   = nn.Linear(1, head_dim)
        self.value_proj = nn.Linear(1, head_dim)
        self.scale = np.sqrt(head_dim)
        self.output_proj = nn.Linear(head_dim, 1)

    def forward(self, x):
        x_reshaped = x.unsqueeze(-1) # (batch_size, num_features, 1)

        q = self.query_proj(x_reshaped)
        k = self.key_proj(x_reshaped)
        v = self.value_proj(x_reshaped)

        attn_scores = torch.softmax((q @ k.transpose(-2, -1)) / self.scale, dim=-1)
        attention_output = attn_scores @ v

        output = self.output_proj(attention_output).squeeze(-1)
        return output


class GatedResidualBlock(nn.Module):
    def __init__(self, input_dim, output_dim, dropout=0.3):
        super().__init__()
        self.linear1 = nn.Linear(input_dim, output_dim)
        self.linear2 = nn.Linear(output_dim, output_dim)
        self.gate = nn.Linear(output_dim, output_dim)
        self.dropout = nn.Dropout(dropout)

        # Using exact nn.Identity mapping for uniform feature dimension paths
        self.skip = nn.Identity() if input_dim == output_dim else nn.Linear(input_dim, output_dim)

    def forward(self, x):
        residual = self.skip(x)
        x_act = torch.relu(self.linear1(x))
        x_act = self.dropout(x_act)
        x_act = self.linear2(x_act)
        x_act = x_act * torch.sigmoid(self.gate(x_act))
        return x_act + residual


# ==========================================
# 3. END-TO-END ADVANCED CLASSIFIER ARCHITECTURE
# ==========================================
class QuantumAttentionInfrastructureGuard(nn.Module):
    def __init__(self, input_dim):
        super().__init__()

        self.attention = HydroFeatureAttention(num_features=input_dim)

        self.encoder = nn.Sequential(
            GatedResidualBlock(input_dim, 32, dropout=0.3),
            GatedResidualBlock(32, 16, dropout=0.3)
        )

        self.to_q1 = nn.Linear(16, n_qubits)
        self.to_q2 = nn.Linear(16, n_qubits)
        self.quantum_dropout = nn.Dropout(p=0.2)

        # Correct shape assignment for our custom two-tier block circuit weights
        # Shape: (2 blocks, q_layers, n_qubits, 3 rotation dimensions)
        weight_shapes = {"weights": (2, q_layers, n_qubits, 3)}
        self.qlayer1 = qml.qnn.TorchLayer(qnode1, weight_shapes)
        self.qlayer2 = qml.qnn.TorchLayer(qnode2, weight_shapes)

        # STABLE UPGRADE: Removed Sigmoid() completely from the final layer block
        self.decoder = nn.Sequential(
            GatedResidualBlock(n_qubits * 2, 16, dropout=0.2),
            nn.Linear(16, 1)
        )

    def forward(self, x):
        x = self.attention(x)
        features = self.encoder(x)

        q1_in = torch.tanh(self.to_q1(features)) * np.pi
        q2_in = torch.tanh(self.to_q2(features)) * np.pi

        q1_in = self.quantum_dropout(q1_in)
        q2_in = self.quantum_dropout(q2_in)

        q1_out = self.qlayer1(q1_in)
        q2_out = self.qlayer2(q2_in)

        combined_quantum_signals = torch.cat((q1_out, q2_out), dim=1)
        return self.decoder(combined_quantum_signals)


# ==========================================
# 4. DATA LOADER PREPARATION
# ==========================================
input_features_count = X_train_raw.shape[1]


# ==========================================
# 5. EXECUTION: TWO-SPEED STABILIZED OPTIMIZER
# ==========================================
model = QuantumAttentionInfrastructureGuard(input_dim=input_features_count)

# HIGH STABILITY LOSS HEADER
criterion = nn.BCEWithLogitsLoss()

classical_params = []
quantum_params = []

for name, param in model.named_parameters():
    if "quantum_layer" in name or "qlayer" in name:
        quantum_params.append(param)
    else:
        classical_params.append(param)

optimizer = optim.AdamW([
    {'params': classical_params, 'lr': 0.005, 'weight_decay': 0.01},
    {'params': quantum_params, 'lr': 0.0005, 'weight_decay': 0.001}
])

epochs = 10
print("🚀 Launching Logit-Stabilized Quantum-Attention Training Loop...\n")

for epoch in range(epochs):
    model.train()
    running_loss = 0.0

    for batch_X, batch_y in train_loader:
        optimizer.zero_grad()

        # Predictions here output raw unbounded logits
        predictions = model(batch_X)
        # Reshape batch_y to match predictions [batch_size, 1]
        loss = criterion(predictions, batch_y.unsqueeze(1).float())

        loss.backward()
        optimizer.step()

        running_loss += loss.item() * batch_X.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    print(f"Epoch [{epoch+1}/{epochs}] - Training Logit BCE Loss: {epoch_loss:.5f}")


# ==========================================
# 6. MODEL VALIDATION / EVALUATION
# ==========================================
model.eval()
test_loss = 0.0
correct_predictions = 0

print("\n🔒 Entering Validation Phase...")
with torch.no_grad():
    for batch_X, batch_y in test_loader:
        test_preds = model(batch_X)
        # Reshape batch_y to match predictions [batch_size, 1]
        loss = criterion(test_preds, batch_y.unsqueeze(1).float())
        test_loss += loss.item() * batch_X.size(0)

        # STABLE ACCURACY CHECK: Explicitly computing the Sigmoid boundary mapping
        binary_preds = (torch.sigmoid(test_preds) >= 0.5).float()
        correct_predictions += (binary_preds == batch_y.unsqueeze(1)).sum().item()

final_test_bce = test_loss / len(test_loader.dataset)
final_accuracy = (correct_predictions / len(test_loader.dataset)) * 100

print(f"➡️ Final Out-of-Sample Test Logit BCE Loss: {final_test_bce:.5f}")
print(f"➡️ Total Classification Accuracy: {final_accuracy:.2f}%")
print("\n✨ Process finished cleanly. Run this exact setup for Burst instances.")
```
